 <font color='cian'>3. Recorro para todas las columnas la tabla y obtengo la cantidad de valores pedidos. Luego construyo una tabla con dicha información y le doy el formato solicitado.</font>

```python
conexion = sqlite3.connect('netflix.db')

columnas = ['id_production', 'id_content', 'title_production', 'genre', 'premiere', 'runtime', 'imdb_score', 'language']

resultados = pd.DataFrame(columns=['Columna', 'NotNull', 'Vacios', 'Null', 'ValoresDistintos'])

for columna in columnas:
    query_notnull = f"SELECT COUNT(*) FROM production WHERE {columna} IS NOT NULL;"
    query_vacios = f"SELECT COUNT(*) FROM production WHERE {columna} = '';"
    query_null = f"SELECT COUNT(*) FROM production WHERE {columna} = '';"
    query_valores_distintos = f"SELECT COUNT(DISTINCT {columna}) FROM production;"
    
    notnull_count = conexion.execute(query_notnull).fetchone()[0]
    vacios_count = conexion.execute(query_vacios).fetchone()[0]
    null_count = conexion.execute(query_null).fetchone()[0]
    valores_distintos = conexion.execute(query_valores_distintos).fetchone()[0]
    
    resultados = resultados.append({'Columna': columna, 'NotNull': notnull_count, 'Vacios': vacios_count, 
    'Null': null_count, 'ValoresDistintos': valores_distintos}, ignore_index=True)

conexion.close()

resultados.rename(columns={'Columna': 'Netflix - production'}, inplace=True)
resultados.set_index('Netflix - production', inplace=True)
resultados_transpose = resultados.transpose()

display(resultados_transpose)
```
