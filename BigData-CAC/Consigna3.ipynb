```python
import pandas as pd

oscar = pd.read_csv('oscar.csv')


########################
import pandas as pd

df = pd.read_csv('oscar.csv')

columnas = list(oscar.columns)

resultados = pd.DataFrame(columns=['Columna', 'NotNull', 'Vacios', 'Null', 'ValoresDistintos'])

for columna in columnas:
    notnull_count = df[columna].notnull().sum()
    vacios_count = (df[columna] == '').sum()
    null_count = df[columna].isnull().sum()
    valores_distintos = df[columna].nunique()
    
    resultados = resultados.append({'Columna': columna, 'NotNull': notnull_count, 'Vacios': vacios_count, 
    'Null': null_count, 'ValoresDistintos': valores_distintos}, ignore_index=True)

resultados.set_index('Columna', inplace=True)

resultados_transpose = resultados.transpose()

display(resultados_transpose)
```
