```python
import folium
import geopandas as gpd
import os

name = 'NSE_VeracruzDeIgnacioDeLaLlave_Alvarado'
file_name = name + '.json'
html_name = name + '.html'
geojson_file = os.path.join(os.curdir, file_name)
gdf = gpd.read_file(geojson_file)

#center = gdf.geometry.unary_union.centroid.coords[:][0]

center = (18.769781339650166, -95.75814359596929) ## Alvarado

## (30.67762960852007, -112.05340214178437) ##NSE_Sonora_Pitiquito

##(27.9108178601378, -110.9090676117093) ## NSE_Sonora_Guaymas

##(25.480424568976517, -107.92170590532595) ## Mocorito

##(22.2162930781896, -100.79927419255743) ## Cerrp san pedro

##(22.352580113351795, -102.88386735785674)  ## Villanueva

##(22.51281730904748, -101.91195668278482) ## Villa Gonz. Ortega
##(23.28754963485798, -102.34574713966288) ## VillaCOs
##(23.636552625868354, -103.63735755387746) ## Sombrerete
##(22.29520461719603, -101.57618628299097) ## Pinos
##(22.877541923305355, -102.53956201560489) ## Panuco


## (24.830593585803452, -101.64166462464264) ## NSE_Zacatecas_Melchor_Ocampo




## (27.290628134694046, -103.7002283506006) ## sierra mojada
##(27.315751613185576, -102.39662042940903) ## Ocampo
## (27.053337004977323, -106.26956740197785) ## EL tule
##(28.239428405757103, -106.83352371898556) # NSE_Chihuahua_Cusihuiriachi
## (31.090249104533907, -107.99632078852287) ## Chihuahua Ascension
## (16.905964740653403, -92.09027814807293) ## CHiapas Ocosingo
## (26.892608890532983, -111.98256895121932) ## BCS Mulege

## (zoom 14)

m = folium.Map(location=center, zoom_start=14) ## ZOOM

##### MUNICIPIOS ENTEROS
    ## Zacatecas(22.81239957141585, -102.60571806622727)
    ## Tamaulipas (23.830889623147552, -99.08240088031957)
    ## Sonora (29.09344872752582, -110.9361418396238)
    ## Sinalao (24.745352976629064, -107.5074244347215)
    ## San luis potosi (22.053465035826928, -100.92535095985873)
    ## Quintana Roo (20.54724509511169, -87.04643933393321)
    ## Queretaro (20.625591544513224, -100.33580548821422)
    ## Nayarit (zoom 13) (21.523783974811796, -104.85476522141106)
    ## Nueva Leon (25.736091920727294, -100.23323126283202)
    ## Jalisco (20.526106276393143, -103.50766099199575)
    ## Guanajuato (20.71956719855507, -101.3358169204181)
    ## Durango (25.570832931193912, -103.45322461081423)
    ## Colima (zoom 11) (19.218292608605946, -104.0654383214647)
    ## Coahuila (26.854310543389015, -101.42351295705616)
    ## Chihuahua (28.68520480399513, -105.98157779629828)
    ## Chiapas (16.754726234865274, -93.12921218625979)
    ## Campeche (19.82391720941923, -90.51235414622683)
    ## CBS (24.143055280263713, -110.29925206615783)
    ## CB (32.485643785554096, -116.87474975518721)  ## COORDENADAS

folium.Marker(location=center).add_to(m)

for i, feature in enumerate(gdf['geometry']):
    clase = gdf['NIVEL PREDOMINANTE'][i]

    clase_color = {'A/B': '#00FF00', 'C+': '#CBFF00', 'C': '#E5FF00', 'C-': '#FFE100', 'D+': '#FFA500', 'D': '#FF8200', 'E': '#FF0000'}
    color = clase_color.get(clase, '#00000000')

    if feature.geom_type == 'Polygon':
        scaled_coordinates = [[y, x] for x, y in zip(feature.exterior.xy[0], feature.exterior.xy[1])]
        folium.Polygon(locations=scaled_coordinates, fill_color=color, fill_opacity=0.7, stroke=False).add_to(m)

    elif feature.geom_type == 'MultiPolygon':
        for polygon in feature.geoms:
            scaled_coordinates = [[y, x] for x, y in zip(polygon.exterior.xy[0], polygon.exterior.xy[1])]
            folium.Polygon(locations=scaled_coordinates, fill_color=color, fill_opacity=0.7, stroke=False).add_to(m)

print(center[1], center[0])

output_html = f'{os.path.splitext(file_name)[0]}.html'
m.save(output_html)
```
